```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("covid_data.csv")

# Convert date
df['date'] = pd.to_datetime(df['date'], errors='coerce')

df = df[(df['date'] >= '2020-01-01') & (df['date'] <= '2022-12-31')]

df = df.dropna(subset=['location', 'total_cases'])
df = df.rename(columns={'location': 'Country'})
```

```python
trend_data = df.sort_values('date').groupby('Country').agg(
    start_cases=('total_cases', 'first'),
    end_cases=('total_cases', 'last')
)

trend_data['change'] = trend_data['end_cases'] - trend_data['start_cases']
trend_data['percent_change'] = (trend_data['change'] / trend_data['start_cases']) * 100

trend_data = trend_data.replace([float('inf'), -float('inf')], 0).dropna()

print(trend_data.head())
```

```python
best_response = trend_data.sort_values('percent_change').head(10)
worst_response = trend_data.sort_values('percent_change', ascending=False).head(10)

print("Best Response Countries:\n", best_response)
print("\nWorst Response Countries:\n", worst_response)
```

```python
plt.figure()
best_response['percent_change'].plot(kind='bar')
plt.title("Best COVID Response (Lowest Growth)")
plt.ylabel("Percent Change")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
worst_response['percent_change'].plot(kind='bar')
plt.title("Worst COVID Response (Highest Growth)")
plt.ylabel("Percent Change")
plt.xticks(rotation=45)
plt.show()
```

```python
top_countries = worst_response.index[:3]

plt.figure()

for country in top_countries:
    data = df[df['Country'] == country]
    plt.plot(data['date'], data['total_cases'], label=country)

plt.title("COVID Case Trends Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.legend()
plt.show()
```

```python
plt.figure()
trend_data['percent_change'].plot(kind='hist', bins=20)
plt.title("Distribution of Case Growth Across Countries")
plt.xlabel("Percent Change")
plt.show()
```
