```python
# NBA Contract Extension Risk Analysis
# Dataset: NBA_Stats_1950_2021_raw.csv

# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv("NBA_Stats_1950_2021_raw.csv")

# Preview Columns
print(df.columns)
```

```python
# CLEAN DATA
# Rename columns if needed depending on dataset
# Example assumes common names:
# Player, Year, Team, G, MP, PTS

# Remove missing values
df = df.dropna()

# Sort by player and season
df = df.sort_values(by=['Player', 'Year'])

# Create Career Season Number (Proxy for Age)
df['Career_Year'] = df.groupby('Player').cumcount() + 1
```

```python
# FIND PLAYER PEAK SEASON
# Peak based on highest points per game

peak_df = df.loc[df.groupby('Player')['PTS'].idxmax()]

avg_peak = peak_df['Career_Year'].mean()
median_peak = peak_df['Career_Year'].median()

print("Average Peak Season:", round(avg_peak,2))
print("Median Peak Season:", median_peak)
```

```python
# VISUALIZATION 1:
# Peak Season Distribution

plt.figure(figsize=(10,6))
sns.histplot(peak_df['Career_Year'], bins=15, kde=True)
plt.title("Distribution of Peak Career Seasons")
plt.xlabel("Career Season")
plt.ylabel("Number of Players")
plt.show()

# VISUALIZATION 2:
# Average Production by Career Year
# -----------------------------
career_avg = df.groupby('Career_Year')['PTS'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=career_avg, x='Career_Year', y='PTS', marker='o')
plt.title("Average Points by Career Year")
plt.xlabel("Career Season")
plt.ylabel("Average Points")
plt.show()
```
