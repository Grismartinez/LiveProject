```python
# Select needed columns
content_df = df[['Category', 'Content Rating', 'Installs']].dropna()

# Group by Content Rating
content_installs = content_df.groupby('Content Rating')['Installs'].sum().sort_values(ascending=False)

print(content_installs)
```

```python
category_count = content_df.groupby('Content Rating')['Category'].nunique().sort_values(ascending=False)

print(category_count)
```

```python
top_content = content_installs.idxmax()
top_installs = content_installs.max()
top_categories = category_count[top_content]

print("Top Content Rating:", top_content)
print("Total Installs:", top_installs)
print("Number of Categories:", top_categories)
```
