```python
df = pd.read_csv("covid_data.csv")
df.head()


# Check columns
df.columns

# Convert date column if needed
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Remove missing values
df = df.dropna(subset=['location', 'new_cases'])

# Rename for clarity (optional)
df = df.rename(columns={'location': 'Country'})
```

```python
# Group by country
country_groups = df.groupby('Country')

avg_cases = df.groupby('Country')['new_cases'].mean().sort_values(ascending=False)

print(avg_cases.head(10))
```

```python
#Visualization

top_countries = avg_cases.head(10)

plt.figure()
top_countries.plot(kind='bar')
plt.title("Top 10 Countries by Average New COVID Cases")
plt.xlabel("Country")
plt.ylabel("Average New Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
selected = avg_cases.head(5)

plt.figure()
selected.plot(kind='bar')
plt.title("Average New Cases Comparison")
plt.xticks(rotation=45)
plt.show()


countries_to_plot = avg_cases.head(3).index

plt.figure()

for country in countries_to_plot:
    country_data = df[df['Country'] == country]
    plt.plot(country_data['date'], country_data['new_cases'], label=country)

plt.title("COVID-19 Trends Over Time")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.legend()
plt.show()
```
